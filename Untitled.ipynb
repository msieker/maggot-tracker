```python
from pymongo import MongoClient
client = MongoClient("mongodb://root:password@localhost:27017")

db = client['maggot-tracker']
tweet_collection = db.tweets
user_collection = db.users
```

```python
import itertools
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, TweetTokenizer
from collections import Counter
from word_cloud.word_cloud_generator import WordCloud
from IPython.core.display import HTML
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
import pandas as pd
import unicodedata
```

```python
tk = TweetTokenizer()
all_stopwords = stopwords.words('english')
all_stopwords += ['!', '.', '[', ']', 'the', 'to', 'rt', ':', '...', '‚Ä¶', ',', '‚Äô', '‚Äú', '‚Äò', '#', '?', 'u', 's', '"', "'", "-", "/", "‚Äù", '(', ')', "&", '*','üá∫', 'üá∏', '@realdonaldtrump', 'q', 'trump']

with open('tagwatch.txt', 'r') as f:
    all_stopwords += [l.strip().lower() for l in f.readlines()]

tweet_text = list(tweet_collection.find(projection={'text': True}))
print('using', len(tweet_text), 'tweets')
cv = CountVectorizer(stop_words=all_stopwords, tokenizer=tk.tokenize, analyzer='word')
word_count_vector = cv.fit_transform([" ".join(list([unicodedata.normalize('NFKD', t['text']).lower() for t in tweet_text]))])
max = np.max(word_count_vector)
word_count_vector = np.multiply(word_count_vector, 1/(max))
coo_matrix = word_count_vector.tocoo()
tuples = zip(coo_matrix.col, coo_matrix.data)
sorted_tuples = sorted(tuples, key=lambda x: (x[1], x[0]), reverse=True)

feature_names = cv.get_feature_names()
counts = [[feature_names[idx], score] for idx, score in sorted_tuples]
df = pd.DataFrame(counts[:100], columns=['words', 'score'])
wc = WordCloud()
embed = wc.get_embed_code(text_scores=df)
HTML(embed)
```
